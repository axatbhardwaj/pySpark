```python
def read_lines(file_path, num_lines):
    with open(file_path, 'r') as file:
        for _ in range(num_lines):
            line = file.readline()
            if not line:
                break
            yield line

def main():
    file_path = '/home/axat/Downloads/new_doc.txt'
    num_lines_to_read = 1000

    lines_generator = read_lines(file_path, num_lines_to_read)

    for line_number, line in enumerate(lines_generator, start=1):
        print(f"Line {line_number}: {line.strip()}")

if __name__ == "__main__":
    main()
```

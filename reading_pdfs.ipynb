```python
import PyPDF2

def replace_text_in_pdf(input_pdf_path, output_pdf_path, text_to_replace, replacement_text):
  """
  Replaces all occurrences of "text_to_replace" with "replacement_text" in each page of the input PDF and saves the modified PDF to "output_pdf_path".

  Args:
    input_pdf_path: The path to the input PDF file.
    output_pdf_path: The path to save the modified PDF file.
    text_to_replace: The text to be replaced.
    replacement_text: The replacement text.
  """

  # Open the input PDF file.
  with open(input_pdf_path, "rb") as f:
    reader = PyPDF2.PdfReader(f)

  # Create a new PDF writer object.
  writer = PyPDF2.PdfWriter()

  # Loop through each page in the input PDF.
  for page in reader.pages:
    # Extract the page content.
    page_content = page.extractText()

    # Replace the text.
    new_page_content = page_content.replace(text_to_replace, replacement_text)

    # Create a new page object and set its content.
    new_page = PyPDF2.PageObject.createBlankPage()
    new_page.drawText(new_page_content)

    # Add the new page to the writer object.
    writer.addPage(new_page)

  # Save the modified PDF.
  with open(output_pdf_path, "wb") as f:
    writer.write(f)

# Example usage
input_pdf_path = "/home/axat/personal/pySpark/data/miri_tech_website_competitors_analysis_detailed.pdf"
output_pdf_path = "/home/axat/personal/pySpark/output/modified_file.pdf"
text_to_replace = "Sumit-wadhwa@india.com"
replacement_text = "robin-saini@india.com"

replace_text_in_pdf(input_pdf_path, output_pdf_path, text_to_replace, replacement_text)

print("Successfully replaced text in PDF.")
```

```python
import fitz
from urllib.parse import urlparse
import re

# Define a function to check if a string is a valid URL
def is_url(s):
    try:
        result = urlparse(s)
        return all([result.scheme, result.netloc])
    except ValueError:
        return False

def process_pdf(file_path):
    doc = fitz.open(file_path)
    new_doc = fitz.open()

    for page_num, page in enumerate(doc):
        text = page.get_text("text")

        # Split text into words and check each for a valid URL
        words = re.findall(r'\S+', text)
        for word in words:
            if is_url(word):
                # Find the coordinates of the word on the page
                word_instances = [match.span() for match in re.finditer(re.escape(word), text)]
                for match in word_instances:
                    start, end = match
                    # Create a rectangle to cover the word
                    rect = fitz.Rect(start, match[0], end, match[1])
                    # You can adjust the color and opacity as needed
                    page.draw_rect(rect, color=(0, 0, 0), fill=(0, 0, 0), overlay=True)
                    print(f"URL Found: {word}")

        new_page = new_doc.new_page(width=page.rect.width, height=page.rect.height)
        new_page.show_pdf_page(page.rect, doc, page_num)

    new_file_path = file_path.replace('.pdf', '_masked.pdf')
    new_doc.save(new_file_path)
    new_doc.close()
    doc.close()
    print(f"Masked PDF saved as {new_file_path}")

file_path = '/home/axat/personal/pySpark/data/blockout.pdf'
process_pdf(file_path)
```

```python
import PyPDF2

def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)

        # Get the number of pages in the PDF
        num_pages = len(pdf_reader.pages)

        # Read text from each page
        text = ''
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()

    return text

# Example usage
file_path = '/home/axat/personal/pySpark/data/EthDev.pdf'
pdf_text = read_pdf(file_path)
print(pdf_text)


#saving that pdf as text file
with open('/home/axat/personal/pySpark/output/EthDev.txt', 'w') as file:
    file.write(pdf_text)

```

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image, ImageDraw, ImageFont
import io

# Function to perform OCR and get text with positions
def ocr_image(image):
    """
    Performs OCR on an image using Tesseract and returns the recognized text along with the positions
    of each text in the image.

    :param image: The image to perform OCR on.
    :return: A list of tuples, where each tuple contains the recognized text and its position
             (x, y, width, height) in the image.
    """
    ocr_result = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)
    text_list = ocr_result['text']
    positions = [(ocr_result['left'][i], ocr_result['top'][i], ocr_result['width'][i], ocr_result['height'][i]) for i in range(len(text_list))]
    return zip(text_list, positions)

# Load PDF
pdf_document = fitz.open('/home/axat/personal/pySpark/data/EthDev.pdf')

# Iterate over each page in the PDF
for page_number in range(pdf_document.page_count):
    # Load the page and get the pixmap
    page = pdf_document.load_page(page_number)
    pix = page.get_pixmap()
    img_data = pix.tobytes("png")
    image = Image.open(io.BytesIO(img_data))

    # Perform OCR on the image
    ocr_results = ocr_image(image)

    # Print OCR results
    print(f"Page {page_number + 1} OCR Results:")
    for text, position in ocr_results:
        print(f"Text: {text}, Position: {position}")

# Close the original PDF
pdf_document.close()
```
